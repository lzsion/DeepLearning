# 第一题

1."人工智能是新电力" 这个比喻指的是什么？

A. 【  】人工智能为我们的家庭和办公室的个人设备供电，类似于电力。

B. 【  】通过“智能电网”，人工智能正在传递新一波的电力。

C. 【  】人工智能在计算机上运行，因此由电力驱动，但它让计算机做以前不可能做的事情。

D. 【  】与100年前开始的电力类似，人工智能正在改变多个行业。

D.【 √ 】与100年前开始的电力类似，人工智能正在改变多个行业。
